```python
# 원본 보호를 위한 복사
hackle_df0_cleaned = hackle_df0.copy()

# 1. 'Asia/Seoul' 컬럼명을 'timestamp'로 변경
hackle_df0_cleaned = hackle_df0_cleaned.rename(columns={'Asia/Seoul': 'timestamp'})

# 2. 문자열 → datetime (UTC 기준으로 변환)
hackle_df0_cleaned['timestamp'] = pd.to_datetime(
    hackle_df0_cleaned['timestamp'], utc=True, errors='coerce'
)

# 3. UTC → KST로 타임존 변환
hackle_df0_cleaned['timestamp_kst'] = hackle_df0_cleaned['timestamp'].dt.tz_convert('Asia/Seoul')

# 4. 날짜(date) 및 시간(time) 컬럼 생성
hackle_df0_cleaned['date'] = hackle_df0_cleaned['timestamp_kst'].dt.date
hackle_df0_cleaned['time'] = hackle_df0_cleaned['timestamp_kst'].dt.time
```

```python
cols = ['timestamp_kst', 'session_id', 'device_id', 'event_key', 'event_properties']
session_df = hackle_df0_cleaned[cols].copy()
```

```python
def extract_session_sequence(df):
    try:
        start_idx = df[df['event_key'] == '$session_start'].index.min()
        end_idx = df[df['event_key'] == '$session_end'].index.max()
        if pd.isna(start_idx) or pd.isna(end_idx):
            return None
        # 해당 구간에서 description이 있는 이벤트만 추출
        sub_df = df.loc[start_idx:end_idx]
        return sub_df['event_description'].dropna().tolist()
    except:
        return None
```

```python
session_sequences = (
    session_df
    .groupby('session_id')
    .apply(extract_session_sequence)
    .dropna()
    .reset_index(name='description_sequence')
)
```
